# **1 -  Aprendizado por Reforço: conceitos, aplicações e desafios**

**O que é Aprendizado por reforço (RL)?**

É um tipo de treinamento de modelos de aprendizado de máquina para tomar uma sequência de decisões. No aprendizado por reforço o agente terá de aprender a atingir uma meta em um ambiente incerto e potencialmente complexo. A máquina irá por tentativa e erro, de acordo as ações que executa a inteligência artificial recebe recompensas ou penalidades. Caracterizado por:

* O feedback é atrasado e ocasional
* O tempo é levando em conta
* As ações do agente afetarão todos os dados subsequentes

**Especialmente útil para tratar de problema de decisão sequencial**

---

O ambiente é geralmente modelado em **processo de decisão de Markov (MDP)**
* **S:** Conjunto de estados em que o processo pode estar
* **A:** Conjunto de ações que podem ser executadas em diferentes épocas de decisão
* **T:** S×A×S → [0, 1] função que dá a probabilidade de o sistema passar para
um estado s' ∈ S, dado que o processo estava em um estado s ∈ S e o agente decidiu executar uma ação a ∈ A (denotada T(s'|s, a))
* **R:** S×A → R é uma função que dá o custo (ou **recompensa**) por tomar uma decisão a ∈ A quando o processo está em um estado s ∈ S.
* **γ:** Fator de desconto que balanceia a troca entre as recompensas imediatas e futuras recompensas, γ∈[0,1]

---

**Relação entre RL e MDP:** O agente aprende a se comportar bem em um abiente modelado em MDV sem saber a função de transição. Faz uma ação e observa a recompensa. Assim o agente aprende a ação que produz a maior soma acumulada das recompensas. **Função (valor, estado, ação)**

→ Solução ótima: π*(s) = arg max Q*(s,a)

---

**Exploration x Exploitation**

* Agente guloso → Agente experiente irá diretamente adquirir a recompensa sem explorar o ambiente (exploit).
* Agente explorador → Agente que irá por tentativa e erro tentar descobrir como adquirir a recompensa, e assim por consequência conhece o ambiente (exploration). **Importante para o aprendizado** (ter certeza de que o ambiente foi explorado) Softmax (distribuição de Boltzmann)

---

**Funcionamento algoritomo Q learning**

  **1** → Inicializa arbitrariamente Q(s,a)

  **2** → Observa o estado em que ele está

  **3** → Até que o critério de parada seja satisfeito, ele selecionará uma ação → executar a ação → receber a recompensa → observar o estado em que ele está depois da execução → atualizar a função Q(s,a)

  Primeiramente o algoritmo passará pela fase de exploração para depois a fase de "exploit".

  * Valor de α alto: exploration
  * Valor de α baixo: exploitation

# **4 - Capítulo 2 do livro: Fundamentos matemáticos do aprendizado por reforço**

### **Componentes do RL:**

→ **O agente: aquele que decide**:

  **Três etapas que o agente faz em um contexto de aprendizado por reforço:** *Avaliar*: todos os agentes avaliam o comportamento do ambiente; *Melhorar*: agentes precisam aprender alguma coisa depois de cada iteração; *Interagir*: os agentes interagem com o problema e com o ambiente.

→ **O ambiente: todo o resto**

Para modelar um problema de processos de tomada de decisão em RL é usando uma estrutura matemática conhecida como processos de decisão de Markov (MDPs)

* O ambiente é representado por um conjunto de variáveis relacionadas ao problema.

Os agentes podem ou não ter acesso ao estado real do ambiente, mas os agentes podem observar algo do ambiente.

* O conjunto de variáveis que o agente percebe em um dado momento é chamado de **observação**.

* A cada estado, o ambiente disponibiliza um conjunto de ações que o agente pode escolher.

**Função de transição:** A função responsável por representar a seguinte relação: O agente tenta influenciar o ambiente através das ações e o ambiente muda ou não de estado como resposta à ação do agente.

* Depois duma transição, o ambiente manda uma nova observação e ainda pode mandar um sinal de recompensa. A função responsável por esse mapeamento é chamada de **função de recompensa**.

**Ambiente Determinístico:** é aquele onde uma ação tem um efeito conhecido, direto e garantido. Não existe incerteza sobre o estado final do ambiente após a execução da ação.

**Ambiente Estocástico:** é quando há uma certa aleatoriedade/imprevisibilidade envolvida na saída da ação.

**Processo que o ambiente passa por causa das ações do agente:** Recebimento da ação selecionada pelo agente → Dependendo do estado atual do ambiente e da ação escolhida pelo agente, o ambiente fará a transição para um novo estado interno → O novo estado e a recompensa passam por um filtro (alguns problemas não deixam o verdadeiro estado do ambiente e não são percebidos pelo agente) → Finalmente, a reação é passada de volta ao agente.

---

### **MDPs: The engine of the environment**

* **Estados:** Configurações específicas do ambiente

O conjunto de todos os estados possíveis, definido como o conjunto **S** ≠ conjunto de variáveis que compõem um único estado.

* **Ações:** Um mecanismo para influenciar o meio ambiente

**A** é uma função que recebe um estado como argumento. Essa função retorna o conjunto de ações disponíveis para o estado s.

O ambiente torna conhecido antecipadamente o conjunto de todas as ações disponíveis. Os agentes podem selecionar ações deterministicamente ou estocasticamente.

Estados terminais: definidos como estados em que todas as ações são deterministicas e transicionam para si mesmos.

* **Função de transição:** Consequências das ações do agente

A função de transição determina a maneira como o ambiente muda em resposta a ações, denotada por T(s, a, s'). Você passa em um estado S, uma ação A e um próximo estado S', e ele retornará a probabilidade correspondente de transição do estado S para o estado S' ao executar a ação A.

* **Sinal de recompensa:** Petisco

A função de recompensa dá um sinal numérico de resposta (o quão bom é resultado) às transições. Quando o sinal é positivo, podemos pensar na recompensa como uma renda ou uma recompensa. Mas, as recompensas também podem ser negativas ativos, e podemos vê-los como custo, punição ou penalidade.

* **Horizonte:** Tempo ideal

Representação do tempo em MDPs, o passo de tempo é conhecido como época, ciclo ou iteração.

Tarefa episódica é uma tarefa em que há um número finito de passos de tempo, seja porque o relógio para ou porque o agente atinge um limite de tempo.

Tarefas contínuas: tarefas que duram para sempre, não há estados terminais, então há um número infinito de passos de tempo. Nesse tipo de tarefa, o agente deve ser parado manualmente.

Horizonte finito: é um horizonte de planejamento no qual o agente sabe que a tarefa terminará em um número finito de passos de tempo.

Horizonte infinito: é quando o agente não tem um limite de passo de tempo predeterminado, então o agente planeja um número infinito de passos de tempo.

* **Desconto:** O futuro é incerto, valorize-o menos

Por causa da possibilidade de infinitas sequências de passos de tempo em tarefas de horizonte infinito, é necessário uma forma de descontar o valor das recompensas ao longo do tempo. Precisamos dizer ao agente que é melhor obter +1 mais cedo ou mais tarde.

Valor de desconto **γ**: Geralmente usa-se um valor real positivo < 1
para descontar exponencialmente o valor das próximas recompensas. Quanto mais longe no futuro recebemos a recompensa, menos valiosa ela é no presente.

### **Função de valor de estado**

É a função que calcula o retorno esperado de um estado, dessa forma é possível comparar diferentes políticas. Define-se o valor de um estado s ao seguir uma certa política π: o valor de um estado s sob a política π é a expectativa de retornos se o agente seguir a política π partindo desse estado s.

Quando calculamos isso para cada estado, obtemos a função de valor de estado, ou função V ou função de valor que representa o retorno esperado ao seguir a política **π** do estado **s**.

**Como funciona?**

**Vπ** = Somatório das ações da política estocástica (soma ponderada) (9) * Somatório da [((ponderamos a soma sobre a) probabilidade das consequência das ações disponíveis no estado atual (10)) * (Retorno do ganho descontado com fator gama (11)], calcule isso para todos estados (12)

$v_{\pi}(s) = \sum_a\pi(a|s) \sum_{s',r}p(s',r|s,a)[r + \gamma v_{\pi} (s')], \quad ∀ s \in S$

### **Função de valor de ação**

Ao comparar diferentes ações sob a mesma política, podemos selecionar as melhores ações e, portanto, melhorar nossas políticas. A função valor-ação, também conhecida como função Q ou Qπ(s,a), captura exatamente isso: o retorno esperado se o agente seguir a política **π** após realizar a ação **a** no estado **s**.

Percebe-se que a função valor de ação é bastante parecida com a função valor de estado. A diferença é que ao invés ponderar sobre todas as ações poderamos somente em uma ação específica.

**Qπ(s,a) (pondera-se a probabilidade das consequência da ação específica escolhida (5))** = Somatório da [((ponderamos a soma sobre a) probabilidade das consequência das ações disponíveis no estado atual (6)) * (Retorno do ganho descontado com fator gama (7))], Devemos fazer isso para todos os pares estado-ação (8)

$q_{\pi}(s,a) = \sum_{s',r}p(s',r|s,a)[r + \gamma v_{\pi} (s')], \quad ∀ s \in S, ∀ a \in A(s)$

### **Função de vantagem de ação**

A função de vantagem de ação, também conhecida como função de vantagem, função Aπ(s, a), é a diferença entre o valor da ação função da ação **a** no estado **s** e a função valor-estado do estado **s** sob a política **π**.

**Aπ(s, a) (1)** = Qπ(s,a) - Vπ(s) (2)

* (1) A vantagem de ação a no estados sob a política π . . .

* (2). . . é a diferença entre o valor dessA ação e o valor da estado s , ambos sob a política π.

$A_{\pi}(s, a) = q_{\pi}(s,a) - V_{\pi}(s)$

A função de vantagem descreve o quanto é melhor tomar a ação **a** em vez de seguir política **π**: a vantagem de escolher a ação **a** sobre a ação padrão.

### **Otimização**

Uma política ótima é uma política que para cada estado pode obter retornos esperados maiores ou iguais a qualquer outra política.

Uma função de valor de estado ótima é uma função de valor de estado com o valor máximo para todos os estados. Da mesma forma, uma função de valor de ação ideal é uma função de valor de ação com o valor máximo para todos os pares de estado-ação.

**V*(s)** = Ação máxima * Somatório da [((ponderamos a soma sobre a) probabilidade das consequência das ações disponíveis no estado atual) * (Retorno do ganho descontado com fator gama]

$v_*(s) = max_a \sum_{s',r}p(s',r|s,a)[r + \gamma v_*(s')]$

$q_*(s,a) = \sum_{s',r}p(s',r|s,a)[r + \gamma max_{a'} q_*(s',a')]$

## **Avaliação da política: políticas de classificação**

O algoritmo de avaliação de políticas consiste em calcular a função V para uma determinada política varrendo o espaço de estados e melhorando iterativamente as estimativas. Referimo-nos ao tipo de algoritmo que recebe uma política e produz uma função de valor como um algoritmo que resolve o problema de previsão, que é calcular os valores de uma política predeterminada.

→ O algoritmo de avaliação de políticas consiste na aproximação iterativa da função estado-valor da política em avaliação:
 * O algoritmo converge quando k se aproxima do infinito.
 * Inicialize **V0(s)** para todos os **s** em **S** arbitrariamente, e para **0** se **s** for terminal. Então, aumente **k** melhore iterativamente as estimativas seguindo a equação abaixo.
 * Calcule o valor de um estado **s** como a soma ponderada da recompensa e o valor estimado descontado do próximo estado **s'**.

```python
# Implementação do algoritmo para avaliação de uma política

#     A função utiliza como parâmetros: pi --> A política que estamos avaliando; P --> O ambiente em MDP;
# gamma --> Fator de desconto de recompensa (=1 Não há desconto); theta --> Número que usamos para checar
# a convergência.
def avaliacao_politica(pi, P, gamma=1.0, theta=1e-10):

  # Inicializa um array com os valores todos zerados e correspondente ao tamanho do ambiente P
  V_anterior = np.zeros(len(P))

  # Loop eterno
  while True:

    V = np.zeros(len(P))

    # Percorrendo todo o array
    for estado in range(len(P)):

      #     O array do ambiente: P[estado][ação da política] --> Em cada posição do array temos uma tupla com
      # as seguintes variaveis que estão sendo percorridas pelo for
      for probabilidade, prox_estado, recompensa, terminado in P[estado][pi(estado)]:

        # Implementação do algoritmo de avaliação de políticas iterativo
        #     A variável terminado é booleana e representa é um estado terminal ou não. Utilizando "not" quando o
        # valor de terminado for "true" o valor cumulativo de V[i] será zerado
        V[estado] += probabilidade * (recompensa + gamma * (V_anterior[prox_estado] * (not terminado)))

    #     Verificamos se a diferença theta é menor que o valor de V. Caso seja menor isso significa que encontramos
    # um valor tão pequenho de diferença que podemos considerar que a política anterior e a atual são iguais
    if np.max(np.abs(V_anterior - V)) < theta:
      # Quebramos o loop infinito
      break

    # Copiamos o V atual para para o V_anterior para continuarmos iterando
    V_anterior = V.copy()

  return V
```

## **Melhoria da política**

Tendo uma maneira de avaliar qualquer política podemos avaliar muitas políticas e classificá-las pela função de valor de estado do estado INICIAL. Afinal, esse número informa a recompensa cumulativa esperada que a política em questão obterá se você executar muitos episódios. Mas, para não gerar aleatoriamente um monte de políticas e avaliá-las. Porque dessa forma seria um desperdício total de recursos de computação, e não teríamos nenhuma garantia de que estamos encontrando políticas cada vez melhores.

Usando função V e o MDP, podemos obter uma estimativa da função Q. A função Q, por sua vez, dará um vislumbre dos valores de todas as ações para todos os estados, e esses valores, por sua vez, podem sugerir como melhorar as políticas.

O algoritmo de **melhoria de política** vai calcular uma função de valor de ação usando a função de valor de estado e o MDP, e retorna uma política **gulosa** em relação a função valor de ação da política original.

→ O algoritmo de melhoria de políticas consiste em usar uma função de valor de estado e um MDP para obter uma antecipação de uma etapa e determinar qual das ações leva ao valor mais alto.
* Obtemos uma nova política π tomando a ação de maior valor.
* Obtemos o valor mais alto de ação calculando, para cada ação, a soma ponderada
de todas as recompensas e valores de todos os próximos estados possíveis.
* Observe que isso usa a ação com a função Q de maior valor (**max**).

```python
# Implementação do algoritmo para melhoria de política

#     A função utiliza como parâmetros: V --> A função valor de estado que estamos melhorando; P --> O ambiente em MDP;
# gamma --> Fator de desconto de recompensa (=1 Não há desconto)
def melhoria_politica(V, P, gamma=1.0):

  #     Inicializa um array com os valores todos zerados e correspondente ao tamanho do ambiente P (len(P))
  # e as ações disponíveis para cada estado do ambiente len(P[0]), todos como valor em ponto flutuante
  Q = np.zeros((len(P), len(P[0])), dtype=np.float64)

  # Percorrendo todo o array
  for estado in range(len(P)):

     # Percorrendo todas as ações disponíveis de cada estado do array
    for acao in range(len(P[estado])):

      #     O array do ambiente: P[estado][ações disponíveis] --> Em cada posição do array temos uma tupla
      # com as seguintes variaveis que estão sendo percorridas pelo for
      for probabilidade, proximo_estado, recompensa, terminado in P[estado][acao]:

        # Implementação do algoritmo de avaliação de políticas iterativo
        #     A variável terminado é booleana e representa é um estado terminal ou não. Utilizando "not" quando o
        # valor de terminado for "true" o valor cumulativo de V[i] será zerado
        #     Obtemos como um resultado o array Q que representa os estados e seus valores para cada ação
        # disponível
        Q[estado][acao] += probabilidade * (recompensa + gamma * V[proximo_estado] * (not terminado))

  #     Implementação do algoritmo de melhoria de políticas
  #     Funções anônimas (Lambda): são  funções que o usuário não precisa definir, ou seja, não vai precisar escrever
  # a função e depois utilizá-la dentro do código.
  #     Criando a política ótima:
  #     A apartir de "np.argmax" o código escolhe qual ação obteve maior valor para cada estado e assim enumera em
  # um plano representada pela tupla dentro da lista, essa lista por sua vez representa a nova política (melhorada)
  new_pi = lambda estado: {estado:acao for estado, acao in enumerate(np.argmax(Q, axis=1))}[estado]

  return new_pi
```

## **Iteração de política**

O plano com essa política adversária é alternar entre avaliação de política e melhoria de política até que a política que sai da fase de melhoria de política não produza mais uma política diferente. O fato é que, se em vez de começarmos com uma política adversária, começarmos com uma política gerada aleatoriamente, é isso que um algoritmo chamado iteração de política faz.

```python

# Implementação do algoritmo para iteração de uma política

#     A função utiliza como parâmetros: P --> O ambiente em MDP; gamma --> Fator de desconto de recompensa
# (=1 Não há desconto); theta --> Número que usamos para checar a convergência.
def policy_iteration(P, gamma=1.0, theta=1e-10):
    #   Note que dessa vez não foi passada uma política como parâmetro. A política é iniciada randomicamente
    # a partir do código abaixo:

    # Criação de ações randomicamente
    acoes_randomicas = np.random.choice(tuple(P[0].keys()), len(P))

    # Criação do array de política randomica
    #     Funções anônimas (Lambda): são  funções que o usuário não precisa definir, ou seja, não vai precisar
    # escrever a função e depois utilizá-la dentro do código.
    pi = lambda estado: {estado:acao for estado, acao in enumerate(acoes_randomicas)}[estado]

    # Loop eterno
    while True:
      # Armazenamos o resultado da política anterior
      pi_antigo = {estado:pi(estado) for estado in range(len(P))}

      # Utilizamos das funções criadas anteriormente
      V = avaliacao_politica(pi, P, gamma, theta)
      pi = melhoria_politica(V, P, gamma)

      #     Verificamos se o valores de avaliação de política já convergiram a aprtir de "theta". Caso sim
      # quebramos o loop eterno.
      if pi_antigo == {estado:pi(estado) for estado in range(len(P))}:
          break

    return V, pi
```

## **Iteração de valor**

Usando uma estimativa truncada da função V e o MDP em cada iteração podemos ver mais facilmente que, mesmo após a primeira iteração, uma política gananciosa sobre as primeiras estimativas da função Q seria uma melhoria.

Em outras palavras, mesmo se truncarmos a avaliação de políticas após uma única iteração, ainda é possível melhorar a política inicial tomando a política gananciosa da estimativa da função Q após uma única varredura no espaço de estado da avaliação de políticas. Este algoritmo é outro algoritmo fundamental
em RL: chama-se iteração de valor (VI).

→ O algoritmo de iteração de valor consiste:
* Mescla-se uma etapa de avaliação de política truncada e uma melhoria de política na mesma equação.
* Calcula-se o valor de cada ação
* E então, tomamos o máximo sobre os valores das ações.
* Multiplique pela probabilidade de cada transição possível.

Observe que na prática, em VI, não temos que lidar com políticas. VI não tem nenhuma fase de avaliação separada que vai até a convergência. Enquanto o objetivo de VI é o mesmo que o objetivo de PI – encontrar a política ótima para um determinado MDP – VI faz isso por meio das funções de valor; assim, a iteração do valor do nome.

```python
# Implementação do algoritmo para Iteração de Valor

#     A função utiliza como parâmetros: P --> O ambiente em MDP; gamma --> Fator de desconto de recompensa
# (=1 Não há desconto); theta --> Número que usamos para checar a convergência.
def value_iteration(P, gamma=1.0, theta=1e-10):

    #     Inicializa um array com os valores todos zerados e correspondente ao tamanho do ambiente P, para
    # armazenar a valor de avaliação da política,  todos como valor em ponto flutuante
    V = np.zeros(len(P), dtype=np.float64)

    # Loop eterno
    while True:
        #     Inicializa um array com os valores todos zerados e correspondente ao tamanho do ambiente P, para
        # armazenar a valor de cada ação de cada estado existente,  todos como valor em ponto flutuante
        Q = np.zeros((len(P), len(P[0])), dtype=np.float64)

        # Percorrendo todos os estados
        for estado in range(len(P)):

            # Percorrendo todos as ações do estados
            for acao in range(len(P[estado])):

                # Percorrendo todos as variáveis da tupla do array
                for probabilidade, proximo_estado, recompensa, terminado in P[estado][acao]:

                    # Implementação do algoritmo de avaliação de políticas iterativo
                    #     A variável terminado é booleana e representa é um estado terminal ou não. Utilizando "not" quando o
                    # valor de terminado for "true" o valor cumulativo de V[i] será zerado
                    #     Obtemos como um resultado o array Q que representa os estados e seus valores para cada ação
                    # disponível
                    Q[estado][acao] += probabilidade * (recompensa + gamma * V[proximo_estado] * (not terminado))

        #     Caso a diferença entre avaliação da política e o valor das maiores ações para política sejam menores que "theta"
        # quebra o loop eterno
        if np.max(np.abs(V - np.max(Q, axis=1))) < theta:
            break

        # Se não guardamos o valor das maiores ações para uma nova política
        V = np.max(Q, axis=1)

    # Terminado o loop eterno temos certeza que obtemos uma política ótima
    pi = lambda estado: {estado:acao for estado, acao in enumerate(np.argmax(Q, axis=1))}[estado]

    return V, pi
```

## **The challenge of interpreting evaluative feedback**

### **Exploitation Vs Exploration**

No capítulo passado nós sabíamos de antemão como o ambiente reagiria a qualquer uma de nossas ações e isso simplifica demais as coisas, de forma até irrealista. Porque isso permite calcular uma política ótima usando algoritmos de planejamento, como PI e VI, sem ter que interagir com o ambiente.

Quando o agente aprende a se comportar a partir da interação com o ambiente, o ambiente faz sempre a mesma pergunta ao agente: o que você quer fazer agora? O agente deve seguir seu conhecimento atual e selecionar a ação com maior estimativa atual? Ou deveria explorar ações que não tentou o suficiente? Ou ainda: Quando você sabe que suas estimativas estão boas o suficiente? Como você sabe que você já tentou uma ação aparentemente ruim o suficiente?

Exploração constrói o conhecimento que permite a exploração efetiva, e a exploração máxima é o objetivo final de qualquer tomador de decisão. É preciso permitir que o agentes interaja e experimente o ambiente por si mesmo, aprendendo assim a se comportar de maneira ideal, por conta de sua própria experiência (tentativa e erro).

---

### **Bandits: problemas de decisão de estado único**

Multi-armed bandits (MAB) são um caso especial de um problema de RL em que temos múltiplas ações, um único estado e um horizonte guloso; você também pode pensar nisso como um ambiente de “muitas opções, uma única escolha”. O nome vem de máquinas caça-níqueis (bandits) com vários braços para escolher.

MABs são MDPs com um único estado não terminal e um único intervalo de tempo por episódio. A função **Q** da ação **a** é a recompensa esperada amostrada. $q(a) = \mathbb{E}[R_t|A_t = a]$

O melhor que podemos fazer em um MAB é representa-lo pela função **V ideal**, ou selecionando a ação que maximiza a função **Q** → $v_* = q(a_*) = max\ q(a)$

A ação ótima é a ação que maximiza a função **Q** ótima, e função **V** ótima (somente um estado) → $a_* = argmax\ q(a)$ ou seja → $q(a_*) = v_*$

---

### **Arrependimento: o custo da exploração**

O objetivo do agente em MABs é o mesmo, ele precisa obter o máximo de recompensa (maximizar) ao longo de um episódio, apesar da estocasticidade do ambiente (esperado). Podemos concluir então que não é um caso “cumulativo” porque há apenas uma única etapa de tempo por episódio, além de que a recompensa nunca será “descontada” porque não há próximos estados a serem contabilizados. Desse modo, no MABs o objetivo é que o agente maximize a recompensa esperada (estocasticidade no ambiente). É por causa da estocasticidade que os agentes do MAB precisam aprender: a distribuição de probabilidade do sinal de recompensa.

A maneira mais robusta de cumprir o objetivo mais completo é o agente maximizar a recompensa esperada por episódio enquanto ainda minimiza a perda total de recompensa esperada de recompensas em todos os episódios. Para calcular esse valor, chamado arrependimento total, somamos a diferença por episódio da verdadeira recompensa esperada da ação ideal e a verdadeira recompensa esperada da ação selecionada (Quanto menor o arrependimento total, melhor). O uso a palavra "verdadeira" remete que você precisa ter acesso ao MDP. Isso não significa que seu agente precisa do MDP, apenas que você precisa dele para comparar a eficiência da estratégia de exploração dos agentes.

$T = \sum \mathbb{E}\ [v_*-q_*(A_e)]$

Para calcular o arrependimento total **T**, é preciso somar, para todos os episódios, a diferença entre o valor ótimo do MAB e o valor real da ação selecionada.

---

### **Abordagens para resolver ambientes MAB**

* **Estratégias de exploração aleatória** → Nosso agente explorará na maior parte do tempo e, às vezes, explorará usando aleatoriedade. Um exemplo dessa estratégia seria um agente que seleciona a ação gulosa na maioria das vezes, e com um limite de épsilon, escolhe uniformemente ao acaso.

* **Estratégias de exploração otimista** → É uma abordadgem mais sistemática que quantifica a incerteza no problema de tomada de decisão e aumenta a preferência por estados com maior incerteza. Essa suposição fará o agente a explorar, porque você assumirá que os estados que ainda não experimentou são os melhores que podem ter.

* **Estratégias de exploração do espaço de estado da informação** → Codificamos a incerteza como parte do espaço de estado significa que um estado do ambiente será visto de forma diferente quando inexplorado ou explorado. Porém ao codificar a incerteza como parte do ambiente podemos aumentar consideravelmente sua complexidade.

Como os MABs são ambientes de uma etapa, para estimar a função **Q**, precisamos apenas calcular a recompensa **média** por **ação**.

---

### **Ganancioso: Sempre guloso**

Para efeito de comparação, vamos considerar primeiramente um algoritmo sem nenhuma exploração. A estratégia **gulosidade pura** (ou estratégia gananciosa) usa uma abordagem de seleção de ação gulosa que consiste em sempre selecionar a ação com o maior valor estimado. Embora haja uma chance da primeira ação que escolhemos ser a melhor ação geral, a probabilidade dessa feliz coincidência diminui à medida que o número de ações disponíveis aumenta.

A estratégia gananciosa fica presa na primeira ação imediatamente, se tabela Q for inicializada em zero e não houver recompensas negativas no ambiente.

* Relação entre uma estratégia gananciosa e o tempo: Se o seu agente tiver apenas um episódio restante, o melhor é agir com ganância. Até certo ponto, é isso que uma estratégia gananciosa faz: ela faz o melhor que pode com sua visão atual da vida, assumindo um tempo limitado.

```python
# Implementação do algoritmo para Pure Exploitation

# A função utiliza como parâmetros: ambiente --> O ambiente em MDP; n_episodios --> Número total episódios para as iterações.
def gulosidade_pura(ambiente, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    # Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos os zerados.

    Q = np.zeros((ambiente.action_space.n)) # Inicializamos a função Q
    N = np.zeros((ambiente.action_space.n)) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, int) # Guarda a ação selecionada a cada iteração

    # Loop principal
    nome = 'Pure exploitation'
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):
        acao = np.argmax(Q) # Seleciona a ação que maximiza os valores Q

        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")
        # Lembrando que como os MABs são ambientes de uma etapa, para estimar a função Q, precisamos apenas calcular a recompensa média por ação.

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"
        
    return nome, retornos, Qe, acoes
```

### **Aleatório: Sempre explorar**

Outra linha de base fundamental que podemos chamar de estratégia aleatória ou estratégia de exploração pura. Esta é simplesmente uma abordagem para seleção de ação sem nenhuma exploração. O único objetivo do agente é obter informações. Lembre-se, embora a exploração seja essencial, ela deve ser bem equilibrada para obter ganhos máximos.

Uma estratégia totalmente aleatória obviamente não é uma boa estratégia e também fornecerá resultados abaixo do ideal. Precisamos de algoritmos que possam fazer "exploitation" e "exploration": obter e usar informações.

```python

# Implementação do algoritmo para Pure Exploration: Aleatório

# A função utiliza como parâmetros: ambiente --> O ambiente em MDP; n_episodios --> Número total episódios para as iterações.
def exploracao_pura(ambiente, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    # Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos os zerados.

    Q = np.zeros((ambiente.action_space.n)) # Inicializamos a função Q
    N = np.zeros((ambiente.action_space.n)) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, int) # Guarda a ação selecionada a cada iteração

    # Loop principal
    nome = 'Pure exploration'
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):
        acao = np.random.randint(len(Q)) # Seleciona uma ação randomicamente

        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")
        # Lembrando que como os MABs são ambientes de uma etapa, para estimar a função Q, precisamos apenas calcular a recompensa média por ação.

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"


    return nome, retornos, Qe, acoes
```

### **Epsilon-ganancioso: Quase sempre ganancioso e às vezes aleatório**

A estratégia epsilon-gananciosa, funciona combinando "exploitation" puro e "exploration" puro, e consiste em agir avidamente na maior parte do tempo e explorar aleatoriamente de vez em quando. Se você selecionar a ação que acha ser a melhor quase o tempo todo, obterá resultados sólidos porque ainda está selecionando a ação que acredita ser a melhor, mas também está selecionando ações que ainda não experimentou o suficiente. Assim, sua função de valor de ação tem a oportunidade de convergir para seu valor verdadeiro e isso ajuda a obter mais recompensas a longo prazo.

Essa estratégia de exploração **continua sendo aleatória**, porque usamos a aleatoriedade para selecionar a ação. Primeiro, usamos a aleatoriedade para escolher entre "exploit" ou "explore", mas também usamos a aleatoriedade para selecionar uma ação exploratória.

Podemos perceber que se o epsilon for 0,5 e você tiver duas ações, você não pode dizer que seu agente irá explorar 50% do tempo, se por “explorar” você quer dizer selecionar a ação não gananciosa. Observe que **a “etapa de exploração” em epsilon-greedy inclui a ação gulosa**. Na realidade, seu agente irá explorar um pouco menos do que o valor do epsilon, dependendo do número de ações.

```python
# Implementação do algoritmo para Epsilon-greedy:

#     A função utiliza como parâmetros: ambiente --> O ambiente em MDP; epsilon --> constante que define uma porcentagem
# para "exploration"; n_episodios --> Número total episódios para as iterações.
def epsilon_ganancioso(ambiente, epsilon=0.01, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    # Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos os zerados.

    Q = np.zeros((ambiente.action_space.n)) # Inicializamos a função Q
    N = np.zeros((ambiente.action_space.n)) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, dtype=int) # Guarda a ação selecionada a cada iteração

    # Loop principal
    nome = f'E-greedy {epsilon}'
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):

        #   É gerado um número randomicamente, se esse número for maior que epsilon
        # o agente agirá gulosamente
        if np.random.random() > epsilon:
            acao = np.argmax(Q) # Seleciona a ação que maximiza os valores Q
        else:
            acao = np.random.randint(len(Q)) # Seleciona uma ação randomicamente

        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"

    return nome, retornos, Qe, acoes
```

```python
nome, retornos, Qe, acoes = epsilon_ganancioso(ambiente, n_episodios=5000)

print(nome)
print(retornos)
print(Qe)
print(acoes)
```

### **Epsilon-ganancioso decadente: primeiro maximize a exploração, depois a exploração**

Se pararmos para pensar o início é quando mais gostaríamos de explorar, quando o agente não experimentou o ambiente o suficiente; depois, à medida que obtém melhores estimativas das funções de valor, queremos que o agente seja gancioso cada vez mais. Assim começamos com um epsilon alto (≤ 1) e diminua seu valor a cada passo.

```python
# Implementação do algoritmo da 1ª Forma algoritmo para Decaying epsilon-greedy: Decaimento linear

#     A função utiliza como parâmetros: ambiente --> O ambiente em MDP; init_epsilon --> constante inicial
# que define uma porcentagem para "exploration"; min_epsilon --> valor mínimo que epsilon pode chegar;
# taxa_decay --> constante que define a taxa de decaimento por episódio; n_episodios --> Número total episódios para
# as iterações.
def lin_dec_epsilon_ganacioso(ambiente, init_epsilon=1.0, min_epsilon=0.01, taxa_decay=0.05, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    # Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos os zerados.

    Q = np.zeros((ambiente.action_space.n)) # Inicializamos a função Q
    N = np.zeros((ambiente.action_space.n)) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, dtype=int) # Guarda a ação selecionada a cada iteração

    # Loop principal
    nome = 'Lin e-greedy {} {} {}'.format(init_epsilon, min_epsilon, taxa_decay)
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):

        decay_episodios = n_episodios * taxa_decay # Número de episódios que gostaríamos de decair epsilon para o valor mínimo

        # Calculos que fazem o valor de epsilon decair linearmente com o aumento de "e"
        epsilon = 1 - e / decay_episodios
        epsilon *= init_epsilon - min_epsilon
        epsilon += min_epsilon
        epsilon = np.clip(epsilon, min_epsilon, init_epsilon)
        #   Numpy funcção clip: Os valores fora do intervalo são recortados nas bordas do intervalo.
        # Por exemplo, se um intervalo de [0, 1] for especificado, os valores menores que 0 tornam-se
        # 0 e os valores maiores que 1 tornam-se 1

        #   É gerado um número randomicamente, se esse número for maior que epsilon
        # o agente agirá gulosamente
        if np.random.random() > epsilon:
            acao = np.argmax(Q) # Seleciona a ação que maximiza os valores Q
        else:
            acao = np.random.randint(len(Q)) # Seleciona uma ação randomicamente

        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"

    return nome, retornos, Qe, acoes
```

```python
nome, retornos, Qe, acoes = lin_dec_epsilon_ganacioso(ambiente, n_episodios=5000)

print(nome)
print(retornos)
print(Qe)
print(acoes)
```

```python
# Implementação do algoritmo da 2ª Forma algoritmo para Decaying epsilon-greedy: Decaimento exponencial

#     A função utiliza como parâmetros: ambiente --> O ambiente em MDP; init_epsilon --> constante inicial
# que define uma porcentagem para "exploration"; min_epsilon --> valor mínimo que epsilon pode chegar;
# taxa_decay --> constante que define a taxa de decaimento por episódio; n_episodios --> Número total episódios para
# as iterações.
def exp_dec_epsilon_ganancioso(ambiente, init_epsilon=1.0, min_epsilon=0.01, taxa_decay=0.1, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    # Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos os zerados.

    Q = np.zeros((ambiente.action_space.n)) # Inicializamos a função Q
    N = np.zeros((ambiente.action_space.n)) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, dtype=int) # Guarda a ação selecionada a cada iteração

    decay_episodios = int(n_episodios * taxa_decay) # Número de episódios que gostaríamos de decair epsilon para o valor mínimo

    # Calculos que fazem o valor de epsilon decair exponencialmente:
    episodios_restantes = n_episodios - decay_episodios
    epsilons = 0.01
    epsilons /= np.logspace(-2, 0, decay_episodios) # Retorna números espaçados uniformemente em uma escala logarítmica.
    # np.logspace(Começo, Final, Número de amostras para gerar)
    epsilons *= init_epsilon - min_epsilon
    epsilons += min_epsilon
    epsilons = np.pad(epsilons, (0, episodios_restantes), 'edge')
    #   Numpy função pad: Preenche um array. np.logspace(O array para preencher,
    # número de valores que são preenchidos nas bordas de cada eixo, Modo EDGE: O preenchimento
    # será feito com os valores de borda)

    # Loop principal
    nome = 'Exp e-greedy {} {} {}'.format(init_epsilon, min_epsilon, taxa_decay)
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):
        #   É gerado um número randomicamente, se esse número for maior que o valor
        # do array epsilon para iteração o agente agirá gulosamente
        if np.random.random() > epsilons[e]:
            acao = np.argmax(Q) # Seleciona a ação que maximiza os valores Q
        else:
            acao = np.random.randint(len(Q)) # Seleciona uma ação randomicamente

        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"

    return nome, retornos, Qe, acoes
```

```python
nome, retornos, Qe, acoes = exp_dec_epsilon_ganancioso(ambiente, n_episodios=5000)

print(nome)
print(retornos)
print(Qe)
print(acoes)
```

### **Inicialização otimista: comece acreditando que é um mundo maravilhoso**

Outra estratégia tratar as ações que você não explorou suficientemente como se fossem as melhores ações possíveis. Essa classe de estratégias é conhecida como otimismo diante da incerteza.

A mecânica dessa estratégia funciona com: primeiramente inicializamos a função Q com um valor alto e depois agimos gulosamente usando essas estimativas. Ao inicializar a função Q com um valor alto, encorajamos a exploração de ações inexploradas. À medida que o agente interage com o ambiente, nossas estimativas começarão a ficar mais baixas, mas mais precisas, permitindo que o agente encontre e converja para a ação com o maior retorno real.

Dois pontos: **“um valor alto”** é algo a que não temos acesso na RL, mas por enquanto, fingiremos que temos esse número com antecedência. O outro é que além dos valores Q, precisamos **inicializar as contagens para um valor maior que um**. Se não fizermos isso, a função Q mudará muito rapidamente e o efeito da estratégia será reduzido.

```python
# Implementação do algoritmo para Optimistic initialization:

#     A função utiliza como parâmetros: ambiente --> O ambiente em MDP; estimativa_otimista --> estimativa
# inicial otimista (valor alto); contagem_inicial --> valor que inicializa as contagens para um valor maior
# que 1; n_episodios --> Número total episódios para as iterações.
def inicializacao_otimista(ambiente, estimativa_otimista=1.0, contagem_inicial=100, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    #   Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos
    # com um valor inicial passados no parâmetro, para que o agente siga de modo otimista.

    Q = np.full((ambiente.action_space.n), estimativa_otimista, dtype=np.float64) # Inicializamos a função Q
    N = np.full((ambiente.action_space.n), contagem_inicial, dtype=np.float64) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, dtype=int) # Guarda a ação selecionada a cada iteração

    # Loop principal
    nome = 'Optimistic {} {}'.format(estimativa_otimista, contagem_inicial)
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):
        acao = np.argmax(Q) # Seleciona a ação que maximiza os valores Q

        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"

    return nome, retornos, Qe, acoes
```

### **Softmax: Selecione ações aleatoriamente em proporção para suas estimativas**

A estratégia softmax amostra uma ação de uma distribuição de probabilidade sobre a função de valor de ação de forma que a probabilidade de selecionar uma ação seja proporcional à sua estimativa de valor de ação atual. Essa estratégia, utiliza da **aleatoriedade na fase de exploração** e está relacionada à estratégia epsilon-greedy. A estratégia do Epsilon-guloso amostra uniformemente e aleatóriamente sobre um conjunto completo de ações disponíveis em um determinado estado, enquanto softmax amostra **baseadas em preferências de ações a de valor mais alto**.

A estratégia softmax, **torna as estimativas de valor de ação em um indicador de preferência**. Não importa quão altos ou baixos sejam os valores; se você adicionar uma constante a todos eles, a distribuição de probabilidade permanecerá a mesma. **Coloca-se preferências sobre a função Q e amostra uma ação de uma distribuição de probabilidade com base nessa preferência**. A diferença entre as estimativas de valor Q criará uma tendência de selecionar ações com as estimativas mais altas com mais frequência e ações com as estimativas mais baixas com menos frequência.

Podemos adicionar um hiperparâmetro para controlar a sensibilidade do algoritmo às diferenças nas estimativas de valor Q. O hiperparâmetro é chamado de **temperatura** e funciona de tal forma que **à medida que se aproxima do infinito, as preferências sobre os valores-Q são iguais**. Basicamente, amostramos uma ação uniformemente. Mas, à medida que o valor da temperatura se aproxima de zero, a ação com o maior valor estimado será amostrada com probabilidade de um. Além disso, podemos decair esse hiperparâmetro linearmente, exponencialmente ou de outra maneira.

Na prática, por causa de estabilidade numérica, **não podemos usar infinito ou zero como temperatura**; em vez disso, usamos um número real positivo muito alto ou muito baixo e normalizamos esses valores.

$\pi(a) = \frac{exp(\frac{Q(a)}{T})}{	\sum_{b=0}^{B} k^2 exp(\frac{Q(b)}{T})}$

* Para calcular a probabilidade de selecionar a ação, calcule a preferência de selecionar essa ação **dividindo** a **função Q** pelo parâmetro de **temperatura T**. Aumente **exponencialmente**. Finalmente, **normalize** os valores **dividindo pela soma de todas as preferências**.

```python
# Implementação do algoritmo para Softmax:

#     A função utiliza como parâmetros: ambiente --> O ambiente em MDP; temperatura_inicial --> temperatura inicial para
# contagem regressiva; temperatura_min --> limite mínimo que a temperatura pode chegar; taxa_decay --> constante que define
# a taxa de decaimento; n_episodios --> Número total episódios para as iterações.
def softmax(ambiente, temperatura_inicial=1000.0, temperatura_min=0.01, taxa_decay=0.04, n_episodios=5000):
    # ambiente.action_space.n --> fornece o número de estados e ação em ambiente carregado
    # Q e N são inicializados com o número de estados e ações condizentes com o ambiente carregado e todos os zerados.

    Q = np.zeros((ambiente.action_space.n)) # Inicializamos a função Q
    N = np.zeros((ambiente.action_space.n)) # Inicializamos a matriz de contagem

    # Declaração de variáveis estatísticas
    Qe = np.empty((n_episodios, ambiente.action_space.n)) # Guarda a função Q a cada episódio (cada iteração)
    retornos = np.empty(n_episodios) # Guarda a recompensa obtida a cada iteração
    acoes = np.empty(n_episodios, dtype=int) # Guarda a ação selecionada a cada iteração

    # Loop principal
    nome = 'SoftMax {} {} {}'.format(temperatura_inicial, temperatura_min, taxa_decay)
    for e in tqdm(range(n_episodios), desc='Episodes for: ' + nome, leave=False):

        decay_episodios = n_episodios * taxa_decay # Número de episódios que gostaríamos de decair a temperatura para o valor mínimo

        # Calculos que fazem o valor da temperatura decair linearmente com o aumento de "e"
        temperatura = 1 - e / decay_episodios
        temperatura *= temperatura_inicial - temperatura_min
        temperatura += temperatura_min
        temperatura = np.clip(temperatura, temperatura_min, temperatura_inicial)
        #   Numpy funcção clip: Os valores fora do intervalo são recortados nas bordas do intervalo.
        # Por exemplo, se um intervalo de [0, 1] for especificado, os valores menores que 0 tornam-se
        # 0 e os valores maiores que 1 tornam-se 1
        #   Usamos a função clip para certificar de que temperatura_min não seja 0, para evitar divisão
        # por zero.

        # Implementação da função softmax (equação acima)
        Q_escalado = Q / temperatura
        Q_normalizado = Q_escalado - np.max(Q_escalado)
        exp_Q = np.exp(Q_normalizado)
        probabilidades = exp_Q / np.sum(exp_Q)

        assert np.isclose(probabilidades.sum(), 1.0) # Verificação de erros com assert
        acao = np.random.choice(np.arange(len(probabilidades)), size=1, p=probabilidades)[0]
        #   Numpy funcção arange: Retorna valores uniformemente espaçados dentro de um
        # determinado intervalo.


        #   ambiente.step() --> Este comando executará uma ação em cada etapa. A ação
        # é especificada como seu parâmetro. A função ambiente.step retorna quatro parâmetros:
        # observação, recompensa, terminado e informações.

        _, recompensa, _, _, _ = ambiente.step(acao) # Rejeitamos observação, terminado e informações com "_" e guardamos o valor de recompensa
        N[acao] += 1 # Adicionamos 1 na matriz de contagem de ações
        Q[acao] = Q[acao] + (recompensa - Q[acao])/N[acao] # Atualizamos a função Q (Mais detalhes em "Melhoria de política")

        # Atualização das estatísticas
        Qe[e] = Q # Guardamos a função Q da iteração "e"
        retornos[e] = recompensa # Guardamos a recompensa da iteração "e"
        acoes[e] = acao # Guardamos a ação selecionada da iteração "e"

    return nome, retornos, Qe, acoes
```

## **Aprendendo a estimar o valor das políticas**

Qual é a melhor forma de avaliar uma política? 

* **Recompensa:** Sinal de recompensa de uma etapa que o agente recebe: o agente observa um estado, seleciona uma ação e recebe um sinal de recompensa. O sinal de recompensa é o núcleo da RL, **mas não é o que o agente está tentando maximizar!** Perceba que, **embora seu agente maximize a recompensa de uma etapa, a longo prazo, ela está recebendo menos do que poderia.**

* **Retorno:** Total de recompensas descontadas. Os retornos são calculados a partir de qualquer estado e costumam ir até o final do episódio. Ou seja, quando um estado terminal é atingido, o cálculo para. Os retornos são frequentemente referidos como recompensa total, recompensa cumulativa, soma de recompensas e são comumente descontados. Como você pode ver, os retornos são melhores indicadores de desempenho porque contêm uma sequência de longo prazo, um histórico de recompensas em um único episódio. **Mas o retorno também não é o que um agente tenta maximizar! Um agente que tenta obter o maior retorno possível pode encontrar uma política que o leve a um caminho ruidoso (muito incerto); às vezes, esse caminho fornecerá um retorno alto, mas talvez na maioria das vezes é baixo.**

* **Função valor:** Expectativa dos retornos. Claro, queremos altos retornos, mas alta expectativa (em média). Se o agente estiver em um ambiente ruidoso ou se estiver usando uma política estocástica, está tudo bem. Afinal, o agente está tentando maximizar a recompensa descontada total esperada: funções de valor.

### **O ambiente Random Walk (RW)**

RW é um ambiente no qual a probabilidade de ir para a esquerda ao executar a ação Esquerda é igual à probabilidade de ir para a direita ao executar a ação Esquerda, e a probabilidade de ir para a direita ao executar a ação Direita ação é igual à probabilidade de ir para a esquerda ao executar a ação para a direita. Em outras palavras, o agente não tem controle de onde ele vai! O agente irá para a esquerda com 50% e para a direita com 50% independente da ação que tomar (aleatório).

Também podemos pensar que RW, ao invés de ser um ambiente sem controle, como um ambiente com uma função de transição determinística (o que significa que, se o agente escolher a esquerda, ele se moverá para a esquerda e se moverá para a direita se escolher a direita - como esperado). Mas suponha que o agente queira avaliar uma política estocástica que seleciona ações uniformemente ao acaso. Isso é, metade do tempo que ele escolhe para a esquerda e a outra metade, para a direita. De qualquer forma, o conceito é o mesmo: temos uma caminhada de cinco estados não terminais em que o agente se move para a esquerda e para a direita uniformemente de forma aleatória. O objetivo é estimar a recompensa descontada total esperada que o agente pode obter dadas essas circunstâncias.

### **Primeira visita Monte Carlo: Melhorando as estimativas a depois de cada episódio**

Vídeo-aula Monte Carlo: https://youtu.be/4xfWzLmIccs

O objetivo é saber quanta recompensa total esperar de uma política. Mais propriamente, o objetivo é estimar a função estado-valor Vπ(s) de uma política π. A abordagem mais direta: é rodar vários episódios com essa política coletando centenas de trajetórias, e depois calcular médias para cada estado, assim como fizemos nos ambientes bandits. Esse método de estimar funções de valor é chamado de previsão de Monte Carlo (MC).

O agente primeiro interagirá com o ambiente usando a política π até que o agente atinja um estado terminal **ST**. A coleção de estado **St**, ação **At**, recompensa **Rt+1** e próximo estado **St+1** é chamada de **tupla de experiência**. **Uma sequência de experiências é chamada de trajetória**. A primeira coisa que você precisa fazer é fazer com que seu agente gere uma trajetória. Depois de ter uma trajetória, você calcula os retornos **Gt:T** para cada estado **St** encontrado, lembrando de descontar as recompensas com **γ**. 

Depois de gerar uma trajetória e calcular os retornos para todos os estados St, você pode estimar a função de valor de estado vπ(s) no final de cada episódio e e o intervalo de tempo final T meramente calculando a média dos retornos obtidos de cada estado s.

**--> Matemática de Monte Carlo**

Função de Valor:
$v_{\pi}(s) = E_{\pi}[G_{t:T}|S_t=s]$

Retornos com valores descontados:
$G_{t:T} = R_{t+1} + γR_{t+2} + y^2R_{t+3} + .... + y^{T-1}R_T$

Dado uma trajetória calcula-se os retornos de cada um dos estados

Depois somamos os retornos de todos os estados:
$T_T(S_t) = T_T(S_t) + G_{t:T}$

Incrementamos uma variável que contará quantas vezes o agente visitou o estado (mais sobre isso na seção "Todas visitas Monte Carlo"):
$N_T(S_t) = N_T(S_t) + 1$

Estimando utilizando a média: 
$V_T(S_t) = \frac{T_T(S_t)}{N_T(S_t)} = V_{T-1}(S_t) + \frac{1}{N_T(S_t)} (G_{t:T} - V_{T-1}(S_t))$

Alpha é igual: $α_t = \frac{1}{N_T(S_t)}$

Finalmente: $V_T(S_t) = V_{T-1}(S_t) + α_t (\begin{matrix} Monte-Carlo-Erro \\ \overbrace{ G_{t:T} - V_{T-1}(S_t)) }  \end{matrix})$

Objetivo: $G_{t:T}$

### **Todas visitas Monte Carlo: Uma maneira diferente de lidar com as visitas de Estado**

Na prática, existem duas maneiras diferentes de implementar um algoritmo de média de retornos. Uma única trajetória pode conter várias visitas ao mesmo estado. Nesse caso, devemos calcular os retornos após cada uma dessas visitas independentemente e depois incluir todas essas metas nas médias ou devemos usar apenas a primeira visita a cada estado?

Ambas são abordagens válidas e têm propriedades teóricas semelhantes. A versão mais “padrão” é o MC de primeira visita (FVMC), e suas propriedades de convergência são fáceis de justificar porque cada trajetória é uma amostra independente e identicamente distribuída (IID) de vπ(s). estimativas convergirão para seus valores verdadeiros. Todas visitas MC (EVMC) é um pouco diferente porque os retornos não são mais independentes e identicamente distribuídos quando os estados são visitados várias vezes na mesma trajetória.

```python
# Função utilizada para decair alfa e epsilon exponencialmente.
# Você dá um valor inicial, um valor mínimo e a porcentagem de max_iteracoes para decair os valores de inicial para mínimo.
def decaimento_exponencial(valor_ini, valor_min, taxa_decaimento, max_iteracoes, log_inicio=-2, log_base=10):
    # Índice onde o decaimento de valores termina e o valor_min vai até max_iteracoes.
    decaimento_iteracoes = int(max_iteracoes * taxa_decaimento)

    # Diferença entre o máximo de iterações e as iterações de decaimento
    iteracoes_remanescentes = max_iteracoes - decaimento_iteracoes

    # Usando o logspace para gerar uma escala logaritimica começando em log_start (padrão = -2), e terminando em 0. 
    # O número total de valores é decaimento_iteracoes e a base é log_base (padrão é 10). 
    # Perceba que está invertido com [::-1]!
    valor = np.logspace(log_inicio, 0, decaimento_iteracoes, base=log_base, endpoint=True)[::-1]

    # Os valores podem não terminar exatamente em 0, por causa do logaritmo
    # Alteramos para ficar entre 0 e 1 e assim a curva fica suave e bonita
    valor = (valor - valor.min()) / (valor.max() - valor.min())

    # Fazemos a transformação linear entre valor_ini e valor_min.
    valor = (valor_ini - valor_min) * valor + valor_min

    # E então enchemos um array com "valor", a função "pad" repete o valor mais à direita iteracoes_remanescentes número de vezes.
    valor = np.pad(valor, (0, iteracoes_remanescentes), 'edge')
    
    return valor
```

```python
# Função utilizada para criar uma trajetória necessária para o método de avaliação de política Monte Carlo
# Você passa a política, o ambiente e a quantidade maxima de iterações (caso o número de iterações ultrapasse 
# este número a trajetória é retornada vazia (truncar)).
def gerar_trajetoria(pi, amb, max_iteracoes=20):
    # Criamos a variavel bolleana "terminado" que somente se tornará True quando o resultado da iteração levar a um estado terminal
    # Inicializamos o array que irá armazenar a trajetória (vazio)
    terminado, trajetoria = False, []

    # Loop principal
    while not terminado:
        # Resetamos o ambiente para garantir que ele esteja pronto para gerar a trajetória
        estado = amb.reset()

        # Entramos em loop infinito -> só para com break
        for t in count():
            # Selecionamos a acao para o estado atual de acordo com a política
            acao = pi(estado)
            # Realizamos uma iteração
            prox_estado, recompensa, terminado, _ = amb.step(acao)
            # Guardamos o resultado da iteração em um array
            experiencia = (estado, acao, recompensa, prox_estado, terminado)
            # append --> Acrescentamos a experiencia no final do array "trajetória"
            trajetoria.append(experiencia)

            # Para o loop quando chegamos em um estado terminal
            if terminado:
                break

            # Truncamos a trajetória quando atinje o máximo de iterações
            if t >= max_iteracoes - 1:
                trajetoria = []
                break

            # Passamos para o proximo estado
            state = prox_estado

    return np.array(trajetoria, np.object)
```

```python
# Implementação do algoritmo para avaliação de política Monte Carlo

# Você passa a política a ser avaliada, o ambiente, desconto de recompensa, especificações para a criação do array de decaimento exponencial de alpha,
# número de episódios considerados, quantidade maxima de iterações e se utilizaremos a primeira vista Monte Carlo ou não
def predicao_mc(pi, amb, gamma=1.0, ini_alpha=0.5, min_alpha=0.01, taxa_decay_alpha=0.3, n_episodios=500, max_iteracoes=100, primeira_visita=True):
    # Obtemos o número de estados do ambiente
    nS = amb.observation_space.n
    
    # Calculando todos os descontos possíveis de uma vez. 
    # Esta função logspace para uma gama de 0,99 e um max_step de 100 retorna um vetor de número 100: [1, 0,99, 0,9801, . . ., 0,3697].
    descontos = np.logspace(0, max_iteracoes, num=max_iteracoes, base=gamma, endpoint=False)
    # Usando a função criada anteriormente para criar o decaimento exponencial de alpha 
    alphas = decaimento_exponencial(ini_alpha, min_alpha, taxa_decay_alpha, n_episodios)
    
    # Inicializando Função de valor
    V = np.zeros(nS)
    # Array utilizado para ver o progresso da avaliação da função de valor
    V_historico = np.zeros((n_episodios, nS))

    # Loop principal com tqdm
    # e --> número do episódio
    for e in tqdm(range(n_episodios), leave=False):

        # Geramos uma trajetoria
        trajetoria = gerar_trajetoria(pi, amb, max_iteracoes)

        # Criamos o array que verifica se a trajetória passou pelo mesmo estado
        visitado = np.zeros(nS, dtype=np.bool)

        # Loop secundário
        # t --> Número da iteração
        # (estado, _, recompensa, _, _) --> Tupla de experiencia da trajetória
        for t, (estado, _, recompensa, _, _) in enumerate(trajetoria):
            # Caso o estado já foi visitado e se estivermos usando a primeira visita MC
            if visitado[estado] and primeira_visita:
                # Ignoramos tudo e vamos para a próxima iteração do loop
                continue

            # Marcamos que estado foi visitado
            visitado[estado] = True

            # Descobrindo a quantidade de passos até o estado terminal
            n_passos = len(trajetoria[t:])
            # Calculando o Retorno: Somando todas recompensas (com seus respectivos descontos) 
            G = np.sum(descontos[:n_passos] * trajetoria[t:, 2])
            # Atualizamos a Função de Valor (Aplicação da equação de Monte Carlo)
            V[estado] = V[estado] + alphas[e] * (G - V[estado])

        # Guardamos a Função de Valor atual no histórico
        V_historico[e] = V
        
    return V.copy(), V_historico
```

### **Aprendizagem da diferença temporal: Melhorando as estimativas depois de cada passo**

Vídeo-aula: https://youtu.be/phgI_880uSw

Uma desvantagem do MC é o fato de que o agente tem que esperar até o final de um episódio, quando pode obter o retorno real Gt:T antes de poder atualizar a estimativa da função estado-valor VT(St) (mesmo tendo propriedades de convergência muito sólidas). Mas, mesmo que esses retornos reais sejam estimativas bastante exatas, eles ao mesmo tempo não são muito precisos. Os retornos reais também são estimativas de alta variância da verdadeira função de valor de estado Vπ(s). É fácil perceber porquê: os retornos reais acumulam muitos eventos aleatórios na mesma trajetória.

Devido à alta variação dos retornos reais Gt:T, MC pode ser ineficiente por amostragem. Toda essa aleatoriedade se torna um ruído que só pode ser atenuado com dados, muitos dados, muitas trajetórias e amostras reais de retorno. Uma maneira de diminuir os problemas de alta variância é, em vez de usar o retorno real Gt:T, estimar um retorno.

Para estimar um retorno podemos usar uma recompensa de etapa única Rt+1 e, depois de observar o próximo estado St+1, pode usar as estimativas da função de valor de estado V(St+1) como uma estimativa do retorno na próxima etapa Gt+ 1:T. Esses métodos, ao contrário do MC, podem aprender a partir de episódios incompletos usando o retorno real de uma etapa, que é a recompensa imediata Rt+1, mas, em seguida, uma estimativa do retorno do próximo estado em diante, que é a estimativa da função valor-estado do próximo estado V(St+1): ou seja, Rt+1 + γV(St+1), que é chamado de alvo TD.

**--> Matemática da Apredendizagem por diferença temporal**

Função de Valor:
$v_{\pi}(s) = E_{\pi}[G_{t:T}|S_t=s]$

Retornos com valores descontados:
$G_{t:T} = R_{t+1} + γR_{t+2} + y^2R_{t+3} + .... + y^{T-1}R_T$

$G_{t:T} = R_{t+1} + γ(R_{t+2} + yR_{t+3} + .... + y^{T-2}R_T)$

$G_{t:T} = R_{t+1} + γG_{t+1:T}$ (Agora $G_{t:T}$ tem estilo recursivo)

Atualizando a Função de Valor:
$v_{\pi}(s) = E_{\pi}[R_{t+1} + γG_{t+1:T}|S_t=s]$ 

(A expectativa de retorno do próximo estado é a função de valor do próximo estado)

Ou seja: $v_{\pi}(s) = E_{\pi}[R_{t+1} + γv_{\pi}(S_{t+1})|S_t=s]$

Assim podemos rodar uma iteração somente e obter $S_t, A_t, R_{t+1}, S_{t+1} -> {\pi}_{t:t+1}$

E assim podemos estimar o valor de $v_{\pi}(s)$

Finalmente: $V_{t+1}(S_t) = V_t(S_t) + α_t (\begin{matrix} Temporal-Difference-erro \\ \overbrace{ R_{t+1} + γV_t(S_{t+1}) - V_t(S_t) }  \end{matrix})$

Objetivo: $R_{t+1} + γV_t(S_{t+1})$

E assim podemos atualizar o valor da estimativa da Função de valor a cada passo de iteração

```python
# Implementação do algoritmo para avaliação de política por diferença temporal

# Você passa a política a ser avaliada, o ambiente, desconto de recompensa, especificações para a criação do array de decaimento exponencial de alpha e
# número de episódios considerados
def td(pi, amb, gamma=1.0, ini_alpha=0.5, min_alpha=0.01, taxa_decay_alpha=0.3, n_episodios=500):
    # Obtemos o número de estados do ambiente
    nS = amb.observation_space.n

    # Inicializando Função de valor
    V = np.zeros(nS)
    # Array utilizado para ver o progresso da avaliação da função de valor
    V_historico = np.zeros((n_episodios, nS))
    # Usando a função criada anteriormente para criar o decaimento exponencial de alpha
    alphas = decaimento_exponencial(ini_alpha, min_alpha, taxa_decay_alpha, n_episodios)

    # Loop principal com tqdm
    # e --> Número da iteração
    for e in tqdm(range(n_episodios), leave=False):
        # Resetamos o ambiente para garantir que ele esteja pronto para ser analisado
        # Obtemos o estado inicial e se o estado é terminal
        estado, terminado = env.reset(), False

        # Loop secundário --> Só para se entrarmos em um estado terminal
        while not done:
            # Selecionamos a acao para o estado atual de acordo com a política
            acao = pi(estado)
            # Obtemos as informações de retorno do passo dado
            prox_estado, recompensa, done, _ = env.step(acao)
            # Implementação da função TD: Calculando o objetivo
            objetivo_td = recompensa + gamma * V[prox_estado] * (not done)
            # Implementação da função TD: Calculando o erro
            erro_td = objetivo_td - V[estado]
            # Atualizamos a Função de Valor (Aplicação da equação de TD)
            V[estado] = V[estado] + alphas[e] * erro_td

            # Passamos para o proximo estado
            state = prox_estado
            
        # Guardamos a Função de Valor atual no histórico
        V_historico[e] = V

    return V, V_historico
```

### **A maioria dos agentes coleta amostras de experiência**

Por último, a maioria dos agentes melhora uma política. ### **Iteração de política generalizada**

No algoritmo de iteração de política (estudado anteriormente), tínhamos 2 processos: avaliação de política e melhoria de política. A fase de avaliação de políticas abrange qualquer política e a avalia; ele estima a função de valor da política. Na melhoria da política, essas estimativas, a função de valor, são usadas para obter uma política melhor. Uma vez que a avaliação e a melhoria da política se estabilizam, a política e a função de valor são ótimas.

O insight crítico aqui é que a avaliação de políticas, em geral, consiste em reunir e estimar funções de valor, semelhantes aos algoritmos que você aprendeu no capítulo anterior.

E a melhoria da política consiste em alterar uma política para torná-la mais gulosa em relação a uma função de valor. No método de melhoria de política do algoritmo de iteração de política, tornamos a política totalmente gulosa em relação à função de valor da política avaliada. Mas, podemos ambicionar completamente a política apenas porque tínhamos o MDP do meio ambiente. No entanto, os métodos de avaliação de políticas que aprendemos no capítulo anterior não exigem um MDP do ambiente, e isso tem um custo. Não podemos mais ambicionar completamente as políticas; precisamos que nossos agentes explorem. No futuro, em vez de ambicionar completamente a política, tornamos a política mais gananciosa, deixando espaço para exploração.

A maioria dos algoritmos RL segue esse padrão de Iteração de política generazilda (GPI): eles têm fases distintas de avaliação e melhoria de políticas, e como existem vários métodos diferentes, tudo o que devemos fazer é escolher o melhor para cada situação dentre eles.

## **Aprendendo a melhorar as políticas de comportamento**

Nesta seção, vamos estudar como resolver o problema de controle: como fazer com que os agentes otimizem as políticas. Vamos utilizar o padrão GPI visto anteriormente e selecionar algoritmos para a fase de avaliação de políticas daqueles que você aprendeu no último capítulo, e estratégias para a fase de melhoria de políticas daqueles que aprendemos no capítulo 4.

* **Problema de predição:** Refere-se ao problema de avaliar políticas, de estimar funções de valor dada uma política (aprender a prever retornos). As funções estado-valor estimam os retornos esperados dos estados e as funções ação-valor estimam os retornos esperados dos pares estado-ação.

* **Problema de controle:** O problema de controle se refere a encontrar políticas ótimas e é resolvido seguindo o padrão GPI. Os métodos de RL geralmente combinam um método de predição de valor de ação com estratégias de melhoria de políticas e seleção de ações.

* **Avaliação de política:** Refere-se ao problema de previsão. Observe que existe um método de programação dinâmica chamado avaliação de política, mas esse termo também é usado para se referir a todos os algoritmos que resolvem o problema de previsão.

* **Melhoria da política:** Refere-se a algoritmos que criam novas políticas que melhoram uma política original, tornando-a mais gananciosa do que a original em relação à função de valor dessa política original. **Observe que a melhoria da política por si só não resolve o problema de controle**. ### **Controle de Monte Carlo: Melhorando as políticas após cada episódio**

Para criar um método de controle usando a previsão de Monte Carlo para nossas necessidades de avaliação de políticas. Não podemos usar a mesma etapa de melhoria de política que usamos para o algoritmo de iteração de política, ou seja, a etapa de melhoria da política obtém a política gulosa em relação à função de valor da política avaliada. Precisamos fazer 2 mudanças que
para fazer funcionar.

Primeiro, precisamos garantir que nosso agente estime o valor da ação
função Q(s, a), em vez de V(s, a) (Lembrando que não é possível saber qual a melhor ação a ser tomada de um estado). E em segundo lugar, precisamos garantir que nosso agente explore (não estamos mais usamos MDP).

Podemos usar a previsão de Monte Carlo na primeira visita para a fase de avaliação de políticas e uma estratégia decadente de seleção de ação gananciosa para a fase de melhoria de políticas.

Parecido com a estratégia da iteração de valor, que tem uma etapa de avaliação de política truncada, podemos truncar o método de previsão de Monte Carlo. Truncamos a etapa de previsão após um único lançamento completo e estimativa de amostra de trajetória e melhoramos a política logo após isso.

Controle MC do método RL em 3 funções:

* decaimento_exponencial: Calcula os valores decrescentes conforme especificado nos argumentos da função.

* gerar_trajetoria_epsilon_ganancioso: implementa a política no ambiente para um episódio completo.

* controle_mc: Implementação completa do método de controle MC.

```python
# Função utilizada para decair alfa e epsilon exponencialmente.
# Você dá um valor inicial, um valor mínimo e a porcentagem de max_iteracoes para decair os valores de inicial para mínimo.
def decaimento_exponencial(valor_ini, valor_min, taxa_decaimento, max_iteracoes, log_inicio=-2, log_base=10):
    # Índice onde o decaimento de valores termina e o valor_min vai até max_iteracoes.
    decaimento_iteracoes = int(max_iteracoes * taxa_decaimento)

    # Diferença entre o máximo de iterações e as iterações de decaimento
    iteracoes_remanescentes = max_iteracoes - decaimento_iteracoes

    # Usando o logspace para gerar uma escala logaritimica começando em log_start (padrão = -2), e terminando em 0. 
    # O número total de valores é decaimento_iteracoes e a base é log_base (padrão é 10). 
    # Perceba que está invertido com [::-1]!
    valor = np.logspace(log_inicio, 0, decaimento_iteracoes, base=log_base, endpoint=True)[::-1]

    # Os valores podem não terminar exatamente em 0, por causa do logaritmo
    # Alteramos para ficar entre 0 e 1 e assim a curva fica suave e bonita
    valor = (valor - valor.min()) / (valor.max() - valor.min())

    # Fazemos a transformação linear entre valor_ini e valor_min.
    valor = (valor_ini - valor_min) * valor + valor_min

    # E então enchemos um array com "valor", a função "pad" repete o valor mais à direita iteracoes_remanescentes número de vezes.
    valor = np.pad(valor, (0, iteracoes_remanescentes), 'edge')
    
    return valor
```

```python
import itertools

# Função utilizada para criar uma trajetória necessária para o controle de Monte Carlo

# Esta versão do gerar_trajetoria é um pouco diferente. Agora precisamos levar em uma estratégia de seleção de ação, em vez de uma política gananciosa.

# Você passa a estratégia de seleção de ação, a função Q, o epsilon (Quão gancioso o agente vai ser?), o ambiente e a quantidade maxima de iterações 
# (caso o número de iterações ultrapasse este número a trajetória é retornada vazia (truncar)).
def gerar_trajetoria_epsilon_ganancioso(seleciona_acao, Q, epsilon, amb, max_iteracoes=200):
    # Criamos a variavel bolleana "terminado" que somente se tornará True quando o resultado da iteração levar a um estado terminal
    # Inicializamos o array que irá armazenar a trajetória (vazio)
    terminado, trajetoria = False, []

    # Loop principal
    while not terminado:
        # Resetamos o ambiente para garantir que ele esteja pronto para gerar a trajetória
        estado = amb.reset()

        # Entramos em loop infinito -> só para com break
        for t in itertools.count():
            # Selecionamos a acao para o estado atual de acordo com a estratégia passada
            acao = seleciona_acao(estado, Q, epsilon)

            # Realizamos uma iteração
            prox_estado, recompensa, terminado, _ = amb.step(acao)
            # Guardamos o resultado da iteração em um array
            experiencia = (estado, acao, recompensa, prox_estado, terminado)
            # append --> Acrescentamos a experiencia no final do array "trajetória"
            trajetoria.append(experiencia)

            # Para o loop quando chegamos em um estado terminal
            if terminado:
                break

            # Truncamos a trajetória quando atinje o máximo de iterações
            if t >= max_iteracoes - 1:
                trajetoria = []
                break

            # Passamos para o proximo estado
            estado = prox_estado
            
        return np.array(trajetoria, np.object)
```

```python
# Implementação do algoritmo de controle de Monte Carlo

# Você passa a política o ambiente para qual o algoritmo criará a política, desconto de recompensa, especificações para a criação do array de 
# decaimento exponencial de alpha e epsilon, número de episódios considerados, quantidade maxima de iterações e se utilizaremos a primeira 
# vista Monte Carlo ou não
def controle_mc(amb, gamma=1.0, ini_alpha=0.5, min_alpha=0.01, taxa_decay_alpha=0.5, ini_epsilon=1.0, min_epsilon=0.1, 
               taxa_decay_epsilon=0.9, n_episodios=3000, max_iteracoes=200, primeira_visita=True):
    # Obtemos o número de estados do ambiente (nS) e o número de ações disponíveis (nA)
    nS, nA = amb.observation_space.n, amb.action_space.n

    # Calculando todos os descontos possíveis de uma vez. 
    # Esta função logspace para uma gama de 0,99 e um max_step de 100 retorna um vetor de número 100: [1, 0,99, 0,9801, . . ., 0,3697].
    descontos = np.logspace(0, max_iteracoes, num=max_iteracoes, base=gamma, endpoint=False)
    # Usando a função criada anteriormente para criar o decaimento exponencial de alpha 
    alphas = decaimento_exponencial(ini_alpha, min_alpha, taxa_decay_alpha, n_episodios)
    # Usando a função criada anteriormente para criar o decaimento exponencial de epsilon 
    epsilons = decaimento_exponencial(ini_epsilon, min_epsilon, taxa_decay_epsilon, n_episodios)

    # Array utilizado para ver o progresso da avaliação da política
    pi_historico = []
    # Inicializando a função com as dimensões do ambiente analisado e os valores zerados
    Q = np.zeros((nS, nA), dtype=np.float64)
    # Array utilizado para ver o progresso da função Q a cada episodio
    Q_historico = np.zeros((n_episodios, nS, nA), dtype=np.float64)

    # Criando a estratégia de seleção de ação: Utilizamos Epsilon Ganancioso
    #     Funções anônimas (Lambda): são  funções que o usuário não precisa definir, ou seja, não vai precisar
    # escrever a função e depois utilizá-la dentro do código.
    seleciona_acao = lambda estado, Q, epsilon: \
        np.argmax(Q[estado]) \
        if np.random.random() > epsilon \
        else np.random.randint(len(Q[estado]))

    # Loop principal com tqdm
    # e --> Número da iteração
    for e in tqdm(range(n_episodios), leave=False):
        # Gerando a trajetória com a função criada anteriormente
        trajetoria = gerar_trajetoria_epsilon_ganancioso(seleciona_acao, Q, epsilons[e], amb, max_iteracoes)

        # Criamos o array que verifica se a trajetória passou pelo mesmo estado
        visitado = np.zeros((nS,nA), dtype=np.bool)

        # Loop secundário
        # t --> Número da iteração
        # (estado, acao, recompensa, _, _) --> Tupla de experiencia da trajetória
        for t, (estado, acao, recompensa, _, _) in enumerate(trajetoria):
            # Caso o estado já foi visitado e se estivermos usando a primeira visita MC
            if visitado[estado][acao] and primeira_visita:
                # Ignoramos tudo e vamos para a próxima iteração do loop
                continue

            # Marcamos que estado foi visitado
            visitado[estado][acao] = True

            # Descobrindo a quantidade de passos até o estado terminal
            n_passos = len(trajetoria[t:])
            # Calculando o Retorno: Somando todas recompensas (com seus respectivos descontos)
            G = np.sum(descontos[:n_passos] * trajetoria[t:, 2])
            # Atualizamos a Função Q (Aplicação da equação de Monte Carlo)
            Q[estado][acao] = Q[estado][acao] + alphas[e] * (G - Q[estado][acao])

        # Guardamos a Função Q atual no histórico
        Q_historico[e] = Q
        # Guardamos a política atual no histórico
        pi_historico.append(np.argmax(Q, axis=1))

    # Extraímos a Função de Valor de Estado selecionando as melhores ações de Q
    V = np.max(Q, axis=1)
    # Com a Função de Valor Feita podemos obter uma política ótima
    pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    return Q, V, pi, Q_historico, pi_historico
```

### **SARSA: Melhorando as políticas após cada passo**

Como estudado no capítulo 5, uma das desvantagens dos métodos de Monte Carlo é que devemos esperar até atingirmos um estado terminal antes de podermos fazer qualquer melhoria em nossas estimativas de função de valor Ao substituir o MC pela previsão TD, agora temos um algoritmo diferente que usa a previsão de diferença temporal para o fase de avaliação de políticas, o conhecido agente SARSA.

```python
# Implementação do algoritmo SARSA

# Você passa a política o ambiente para qual o algoritmo criará a política, desconto de recompensa, especificações para a criação do array de 
# decaimento exponencial de alpha e epsilon, número de episódios considerados
def sarsa(amb, gamma=1.0, ini_alpha=0.5, min_alpha=0.01, taxa_decay_alpha=0.5, ini_epsilon=1.0, 
          min_epsilon=0.1, taxa_decay_epsilon=0.9, n_episodios=3000):
    # Obtemos o número de estados do ambiente (nS) e o número de ações disponíveis (nA)
    nS, nA = env.observation_space.n, env.action_space.n

    # Array utilizado para ver o progresso da avaliação da política
    pi_historico = []
    # Inicializando a função com as dimensões do ambiente analisado e os valores zerados
    Q = np.zeros((nS, nA), dtype=np.float64)
    # Array utilizado para ver o progresso da função Q a cada episodio
    Q_historico = np.zeros((n_episodios, nS, nA), dtype=np.float64)

    # Criando a estratégia de seleção de ação: Utilizamos Epsilon Ganancioso
    #     Funções anônimas (Lambda): são  funções que o usuário não precisa definir, ou seja, não vai precisar
    # escrever a função e depois utilizá-la dentro do código.
    seleciona_acao = lambda estado, Q, epsilon: \
        np.argmax(Q[estado]) \
        if np.random.random() > epsilon \
        else np.random.randint(len(Q[estado]))

    # Usando a função criada anteriormente para criar o decaimento exponencial de alpha 
    alphas = decaimento_exponencial(ini_alpha, min_alpha, taxa_decay_alpha, n_episodios)
    # Usando a função criada anteriormente para criar o decaimento exponencial de epsilon 
    epsilons = decaimento_exponencial(ini_epsilon, min_epsilon, taxa_decay_epsilon, n_episodios)

    # Loop principal com tqdm
    # e --> Número da iteração
    for e in tqdm(range(n_episodios), leave=False):
        # Resetamos o ambiente para garantir que ele esteja pronto para ser analisado
        # Obtemos o estado inicial e se o estado é terminal
        estado, terminado = amb.reset(), False
        # Selecionamos a acao para o estado atual de acordo com a estratégia passada
        acao = seleciona_acao(estado, Q, epsilons[e])
        
        # Loop secundário --> Só para se entrarmos em um estado terminal
        while not terminado:
            # Obtemos as informações de retorno do passo dado
            prox_estado, recompensa, terminado, _ = amb.step(acao)

            # Selecionamos a acao para o estado atual de acordo com a estratégia passada
            # Observe que antes de fazer qualquer cálculo, precisamos obter a ação para a próxima etapa.
            prox_acao = seleciona_acao(prox_estado, Q, epsilons[e])

            # Implementação da função TD: Calculando o objetivo (Caso o estado seja terminal temos que zerar)
            objetivo_td = recompensa + gamma * Q[prox_estado][prox_acao] * (not terminado)
            # Implementação da função TD: Calculando o erro
            erro_td = objetivo_td - Q[estado][acao]
            # Implementação da função TD: Calculando a função Q
            Q[estado][acao] = Q[estado][acao] + alphas[e] * erro_td

            # Passamos para o proximo estado e proxima ação
            estado, acao = prox_estado, prox_acao

        # Guardamos a Função Q atual no histórico
        Q_historico[e] = Q
        # Guardamos a política atual no histórico
        pi_historico.append(np.argmax(Q, axis=1))

    # Extraímos a Função de Valor de Estado selecionando as melhores ações de Q
    V = np.max(Q, axis=1)
    # Com a Função de Valor Feita podemos obter uma política ótima
    pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    return Q, V, pi, Q_historico, pi_historico
```

## **Separando o comportamento da aprendizagem**

### **Q-learning: Aprender a agir de forma otimizada, mesmo que optemos por não fazê-lo**

O algoritmo SARSA é uma espécie de “aprendizagem no trabalho”. O agente aprende sobre a mesma política que usa para gerar experiência (**aprendizado dentro da política**, aprendendo com os próprios erros). No aprendizado sobre políticas, aprendemos apenas com nossos próprios erros atuais. Para aprender com nossos próprios erros anteriores e com os erros dos outros utilizamos a **aprendizagem fora da política**.

Nesse tipo de aprendizagem o agente aprende sobre uma política que é diferente das experiências de geração de políticas. No aprendizado off-policy, existem 2 políticas: uma política de comportamento, usada para gerar experiências, para interagir com o ambiente, e uma política de destino, que é a política sobre a qual estamos aprendendo. O SARSA é um método on-policy; O Q-learning está fora da política.

O avançado algoritmo Q-learning utiliza essas possibilidades, é um método de **bootstrapping fora da política** sem modelo que se aproxima diretamente da política ideal, apesar das experiências de geração de política. Isso significa que o agente, em teoria, pode agir aleatoriamente e ainda encontrar a função de valor e as políticas ótimas.

**--> Matemática: SARSA Vs Q-learning**

Essa é a equação de atualização do **SARSA**:
$Q(S_t, A_t) 	\leftarrow Q(S_t, A_t) + α_t (\begin{matrix} SARSA-erro \\ \overbrace{ R_{t+1} + γQ_t(S_{t+1}, A_{t+1}) - Q(S_t, A_t) }  \end{matrix})$

Objetivo do **SARSA**: $R_{t+1} + γQ_t(S_{t+1}, A_{t+1})$

Ela utiliza a ação escolhida no próximo estado para calcular o objetivo

E essa é a equação de atualização do **Q-learning**:
$Q(S_t, A_t) 	\leftarrow Q(S_t, A_t) + α_t (\begin{matrix} Q-learning-erro \\ \overbrace{ R_{t+1} + γmax_aQ(S_{t+1}, a) - Q(S_t, A_t) }  \end{matrix})$

Objetivo do **Q-learning**: $R_{t+1} + γmax_aQ(S_{t+1}, a)$

O Q-learning usa a ação com o valor máximo estimado no próximo estado, apesar da ação executada.

```python
# Implementação do algoritmo Q-Learning

# Você passa a política o ambiente para qual o algoritmo criará a política, desconto de recompensa, especificações para a criação do array de 
# decaimento exponencial de alpha e epsilon, número de episódios considerados
def q_learning(amb, gamma=1.0, ini_alpha=0.5, min_alpha=0.01, taxa_decay_alpha=0.5, ini_epsilon=1.0, 
          min_epsilon=0.1, taxa_decay_epsilon=0.9, n_episodios=3000):
    # Obtemos o número de estados do ambiente (nS) e o número de ações disponíveis (nA)
    nS, nA = env.observation_space.n, env.action_space.n

    # Array utilizado para ver o progresso da avaliação da política
    pi_historico = []
    # Inicializando a função com as dimensões do ambiente analisado e os valores zerados
    Q = np.zeros((nS, nA), dtype=np.float64)
    # Array utilizado para ver o progresso da função Q a cada episodio
    Q_historico = np.zeros((n_episodios, nS, nA), dtype=np.float64)

    # Criando a estratégia de seleção de ação: Utilizamos Epsilon Ganancioso
    #     Funções anônimas (Lambda): são  funções que o usuário não precisa definir, ou seja, não vai precisar
    # escrever a função e depois utilizá-la dentro do código.
    seleciona_acao = lambda estado, Q, epsilon: \
        np.argmax(Q[estado]) \
        if np.random.random() > epsilon \
        else np.random.randint(len(Q[estado]))

    # Usando a função criada anteriormente para criar o decaimento exponencial de alpha 
    alphas = decaimento_exponencial(ini_alpha, min_alpha, taxa_decay_alpha, n_episodios)
    # Usando a função criada anteriormente para criar o decaimento exponencial de epsilon 
    epsilons = decaimento_exponencial(ini_epsilon, min_epsilon, taxa_decay_epsilon, n_episodios)

    # Loop principal com tqdm
    # e --> Número da iteração
    for e in tqdm(range(n_episodios), leave=False):
        # Resetamos o ambiente para garantir que ele esteja pronto para ser analisado
        # Obtemos o estado inicial e se o estado é terminal
        estado, terminado = amb.reset(), False

        # Loop secundário --> Só para se entrarmos em um estado terminal
        while not terminado:

            # Selecionamos a acao para o estado atual de acordo com a estratégia passada
            acao = seleciona_acao(estado, Q, epsilons[e])
            # Obtemos as informações de retorno do passo dado
            prox_estado, recompensa, terminado, _ = amb.step(acao)

            # Implementação da função Q-learning: Calculando o objetivo (Caso o estado seja terminal temos que zerar)
            objetivo_td = recompensa + gamma * Q[prox_estado].max() * (not terminado)
            # Implementação da função Q-learning: Calculando o erro
            erro_td = objetivo_td - Q[estado][acao]
            # Implementação da função Q-learning: Calculando a Função Q
            Q[estado][acao] = Q[estado][acao] + alphas[e] * erro_td

            # Passamos para o proximo estado
            estado = prox_estado      

        # Guardamos a Função Q atual no histórico
        Q_historico[e] = Q
        # Guardamos a política atual no histórico
        pi_historico.append(np.argmax(Q, axis=1))

    # Extraímos a Função de Valor de Estado selecionando as melhores ações de Q
    V = np.max(Q, axis=1)
    # Com a Função de Valor Feita podemos obter uma política ótima
    pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    return Q, V, pi, Q_historico, pi_historico
```

* **On-policy learning:** Refere-se a métodos que tentam avaliar ou melhorar a política usada para tomar decisões. Esta política gera comportamento. Seu agente avalia esse comportamento e seleciona áreas de melhoria com base nessas estimativas. Seu agente aprende a avaliar e melhorar a mesma política que usa para gerar os dados.

Seu agente aprende a avaliar e melhorar uma política diferente daquela usada para gerar os dados.

### **Q-learning duplo: o máximo de estimativas para uma estimativa de um máximo**

O Q-learning geralmente superestima a função de valor. Em cada passo, tomamos o máximo sobre as estimativas da função valor-ação do próximo estado. Mas o que precisamos é o **valor real** da função máxima de valor de ação do próximo estado. **Em outras palavras, estamos usando o máximo sobre apenas estimativas como uma estimativa do máximo.**

Isso é uma maneira imprecisa de estimar o valor máximo e ainda um problema pior, essas estimativas de bootstrap que são usadas para formar alvos de TD, porém na maioria das vezes são tendenciosas.

A grande diferença para o Q learning normal está em utilizar dois valores de funções-estado Q1 e Q2. Você pode pensar nisso de forma semelhante à validação cruzada: as estimativas de uma função Q ajuda a validar as outras estimativas da outra função Q. O problema, porém, é que agora estamos dividindo a experiência entre duas funções separadas. Isso atrasa um pouco o treinamento.

```python
# Implementação do algoritmo Q-Learning duplo

# Você passa a política o ambiente para qual o algoritmo criará a política, desconto de recompensa, especificações para a criação do array de 
# decaimento exponencial de alpha e epsilon, número de episódios considerados
def double_q_learning(amb, gamma=1.0, ini_alpha=0.5, min_alpha=0.01, taxa_decay_alpha=0.5, ini_epsilon=1.0,
                      min_epsilon=0.1, taxa_decay_epsilon=0.9, n_episodios=3000):
    # Obtemos o número de estados do ambiente (nS) e o número de ações disponíveis (nA)
    nS, nA = amb.observation_space.n, amb.action_space.n
    
    # Array utilizado para ver o progresso da avaliação da política
    pi_historico = []

     # Inicializando a função Q1 e Q2 com as dimensões do ambiente analisado e os valores zerados
    Q1 = np.zeros((nS, nA), dtype=np.float64)
    Q2 = np.zeros((nS, nA), dtype=np.float64)
    # Arrays utilizados para ver o progresso da funções Q1 e Q2 a cada episodio
    Q1_historico = np.zeros((n_episodios, nS, nA), dtype=np.float64)
    Q2_historico = np.zeros((n_episodios, nS, nA), dtype=np.float64)

    # Criando a estratégia de seleção de ação: Utilizamos Epsilon Ganancioso
    #     Funções anônimas (Lambda): são  funções que o usuário não precisa definir, ou seja, não vai precisar
    # escrever a função e depois utilizá-la dentro do código.
    seleciona_acao = lambda estado, Q, epsilon: \
        np.argmax(Q[estado]) \
        if np.random.random() > epsilon \
        else np.random.randint(len(Q[estado]))

    # Usando a função criada anteriormente para criar o decaimento exponencial de alpha 
    alphas = decaimento_exponencial(ini_alpha, min_alpha, taxa_decay_alpha, n_episodios)
    # Usando a função criada anteriormente para criar o decaimento exponencial de epsilon 
    epsilons = decaimento_exponencial(ini_epsilon, min_epsilon, taxa_decay_epsilon, n_episodios)

    # Loop principal com tqdm
    # e --> Número da iteração
    for e in tqdm(range(n_episodios), leave=False):
        # Resetamos o ambiente para garantir que ele esteja pronto para ser analisado
        # Obtemos o estado inicial e se o estado é terminal
        estado, terminado = amb.reset(), False

        # Loop secundário --> Só para se entrarmos em um estado terminal
        while not terminado:
            # Selecionamos a acao para o estado atual de acordo com a estratégia passada
            # Precisamos passar somente uma função Q por isso criamos um array com a média de nossas duas funções Q
            # Também poderíamos usar a soma de nossas funções Q aqui, eles darão resultados semelhantes.
            acao = seleciona_acao(estado, (Q1 + Q2)/2., epsilons[e])

            # Obtemos as informações de retorno do passo dado
            prox_estado, recompensa, terminado, _ = amb.step(acao)

            # Escolhemos aleatoriamente (50%) qual função Q usar
            if np.random.randint(2):
                # Vamos usar a ação que Q1 acha melhor
                argmax_Q1 = np.argmax(Q1[prox_estado])

                # Mas usamos o valor de Q2 para calcular o objetivo TD
                objetivo_td = recompensa + gamma * Q2[prox_estado][argmax_Q1] * (not terminado)
                # Resto da implementação da função do Q-learning: Usando Q1
                erro_td = objetivo_td - Q1[estado][acao]
                Q1[estado][acao] = Q1[estado][acao] + alphas[e] * erro_td

            else:
                # Vamos usar a ação que Q2 acha melhor
                argmax_Q2 = np.argmax(Q2[prox_estado])

                # Mas usamos o valor de Q2 para calcular o objetivo TD
                objetivo_td = recompensa + gamma * Q1[prox_estado][argmax_Q2] * (not terminado)
                # Resto da implementação da função do Q-learning: Usando Q2
                erro_td = objetivo_td - Q2[estado][acao]
                Q2[estado][acao] = Q2[estado][acao] + alphas[e] * erro_td

            # Passamos para o proximo estado
            estado = prox_estado

        # Guardamos as Funções Q1 e Q2 atuais no histórico
        Q1_historico[e] = Q1
        Q2_historico[e] = Q2
        # Guardamos a política atual no histórico, perceba que utilizaremos a avaliação média entre Q1 e Q2
        pi_historico.append(np.argmax((Q1 + Q2)/2., axis=1))

    # O Q final é média do 1 com o 2
    Q = (Q1 + Q2)/2.
    # Obtemos a Função valor estado
    V = np.max(Q, axis=1)
    # Obtemos uma política ótima
    pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    # Perceba que retornamos o histórico médio de Q
    return Q, V, pi, (Q1_historico + Q2_historico)/2., pi_historico
```

```python
# Valores ótimos -> Para comparação

# Iteração de valor do capítulo 3
def iteracao_de_valor(P, gamma=1.0, theta=1e-10):
    V = np.zeros(len(P), dtype=np.float64)
    while True:
        Q = np.zeros((len(P), len(P[0])), dtype=np.float64)
        for s in range(len(P)):
            for a in range(len(P[s])):
                for prob, next_state, reward, done in P[s][a]:
                    Q[s][a] += prob * (reward + gamma * V[next_state] * (not done))
        if np.max(np.abs(V - np.max(Q, axis=1))) < theta:
            break
        V = np.max(Q, axis=1)
    pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    return Q, V, pi

amb = old_gym.make('SlipperyWalkSeven-v0')
amb.reset()
Q_otimo, V_otimo, pi_otimo = iteracao_de_valor(P, gamma=gamma)

print("-----------Valores ótimos------------")

imprima_funcao_valor_de_estado(V_otimo, P, title="\n\nFunção V (valor-estado) ótima")
imprima_funcao_valor_de_acao(Q_otimo, title="\n\n\nFunção Q (valor-ação) ótima")
imprima_politica(pi_otimo, P, title="\n\nPolítica ótima")
```

```python
from itertools import cycle

def plotar_funcao_de_valor(title, V_track, V_true=None, log=False, limit_value=0.05, limit_items=5):
    np.random.seed(123)
    per_col = 25
    linecycler = cycle(["-","--",":","-."])
    legends = []

    valid_values = np.argwhere(V_track[-1] > limit_value).squeeze()
    items_idxs = np.random.choice(valid_values, 
                                  min(valid_values.size, limit_items), 
                                  replace=False)
    # draw the true values first
    if V_true is not None:
        for i, state in enumerate(V_track.T):
            if i not in items_idxs:
                continue
            if state[-1] < limit_value:
                continue

            label = 'v*({})'.format(i)
            plt.axhline(y=V_true[i], color='k', linestyle='-', linewidth=1)
            plt.text(int(len(V_track)*1.02), V_true[i]+.01, label)

    # then the estimates
    for i, state in enumerate(V_track.T):
        if i not in items_idxs:
            continue
        if state[-1] < limit_value:
            continue
        line_type = next(linecycler)
        label = 'V({})'.format(i)
        p, = plt.plot(state, line_type, label=label, linewidth=3)
        legends.append(p)
        
    legends.reverse()

    ls = []
    for loc, idx in enumerate(range(0, len(legends), per_col)):
        subset = legends[idx:idx+per_col]
        l = plt.legend(subset, [p.get_label() for p in subset], 
                       loc='center right', bbox_to_anchor=(1.25, 0.5))
        ls.append(l)
    [plt.gca().add_artist(l) for l in ls[:-1]]
    if log: plt.xscale('log')
    plt.title(title)
    plt.ylabel('State-value function')
    plt.xlabel('Episodes (log scale)' if log else 'Episodes')
    plt.show()
```

### **Considerações Finais:**
* Para a estrtégia de **Monte Carlo**: as estimativas **têm alta variação**, como no algoritmo de previsão portanto **não possuem uma boa precisão**. E **leva** uma boa quantidade de **tempo** para chegar aos valores ideais.

* Para a estrtégia de **SARSA**: Nos testes, podemos ver que ele tem **menos variação do que o MC**, **mas leva praticamente a mesma quantidade de tempo** para chegar aos valores ideais.

* Para a estrtégia de **Q-learning**: Essa estratégia é um método de inicialização fora da política, dessa forma, é **muito mais rapida para chegar as estimativas com os valores ideais**. Mas, também, observe como as **estimativas** costumam ser mais **altas** e saltam de forma um tanto agressiva.

* Para a estrtégia de **Q-learning duplo**: é um **pouco mais lento do que o Q-learning** para obter as estimativas da função de valor de estado ideal, mas é muito **mais estável**. Ainda exite um pouco de superestimação, mas está controlada.
